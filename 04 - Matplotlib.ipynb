# Matplotlib

````python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
````

````python
x = np.linspace(0,5,10)
y = x ** 2
````

````python
plt.plot(x,y)
plt.title("Parabola Test")
plt.xlabel("x axis")
plt.ylabel("y axis")

#not needed
plt.show()
````

## subplot

````python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(x,y,'g')
````

## l'oggetto figure

````python
fig1 = plt.figure(figsize=(5,4),dpi=150)
#percent of figure that want we use
ax1 = fig1.add_axes([0.1,0.1,0.9,0.9])

ax1.set_xlabel("x axis")
ax1.set_ylabel("y axis")
ax1.plot(x,y,label="parabola")
ax1.plot(y,x,label="parabola inversed")
ax1.legend()

ax2 = fig1.add_axes([0.45,0.45,0.4,0.4])
ax2.plot(x,x,'r',label="line")
#ax1.text(0,40,"Message")
````

## Subplots

````python
fig,axis = plt.subplots(figsize=(8,4),nrows=1,ncols=3)
#aggiusta automaticamente gli spazi tra i grafici
plt.tight_layout()

axis[1].set_title("Plot 2")
axis[1].set_xlabel("x axis")
axis[1].set_ylabel("y axis")
axis[1].plot(x,y,label="parabola")
````

## impostazioni grafiche

i colori sono definiti nei seguenti modi:
* b blue
* g green
* r red
* c cyan
* m magenta
* y yellow
* k black
* w white

se impostiamo ```color=0.75``` impostiamo il grigio al 75% possiamo anche usare i colori in esadecimale ad esempio ```color=#eeefff``` oppure usare i nomi dei colori presi dalla seguente [tabellina](https://en.wikipedia.org/wiki/Web_colors)

Possiamo anche impostare lo stile della linea secondo questo link di [esempio](https://matplotlib.org/3.1.0/gallery/lines_bars_and_markers/linestyles.html)

I markers invece sono i punti in cui la linea passa per scegliere il loro stile si può seguire questo [esempio](https://matplotlib.org/3.3.3/api/markers_api.html)

````python
plt.plot(x,y,color="navy",alpha=0.75,lw = 3,ls="-.",marker="o",markersize=7)
````

````python
fig1 = plt.figure(figsize=(5,4),dpi=150)
ax1 = fig1.add_axes([0.1,0.1,0.9,0.9])
ax1.set_xlim([0,3])
ax1.set_ylim([0,8])
ax1.plot(x,y,color="navy",alpha=0.75,lw = 3,ls="-.",marker="o",markersize=7)
ax1.grid(True,dashes=(5,2,1,2),color=".6")
ax1.set_facecolor("#faebd7")
````

## lavorare con pandas

````python
import datetime
google_df = pd.read_csv('./data/google_stocks.csv',parse_dates = ["Date"])
google_df = google_df.sort_values('Date')

x = google_df['Date']
y = google_df['Close']

fig = plt.figure(figsize=(5,4),dpi=150)
ax = fig.add_axes([0.1,0.1,0.9,0.9])

ax.set_title("Google stocks")
ax.set_xlabel("Date")
ax.set_ylabel("Close")
ax.set_ylim([0,1400])
ax.set_xlim([datetime.date(2004, 1, 1), datetime.date(2019, 12, 31)])
ax.plot_date(x, y, marker='', linestyle='-',color="g")
ax.grid(True,dashes=(5,2,1,2),color=".6")
fig.autofmt_xdate()

ax.annotate("Best value 1287.58 $",
            xy=(datetime.date(2019, 4, 29),1287.58),
            xytext=(datetime.date(2010, 1, 1),1260),
            arrowprops=dict(facecolor="black",shrink=0.05)
           )
````

## Grafico a torta

````python
import random

fig = plt.figure(figsize=(8,5),dpi=100)
axes = fig.add_axes([0.1,0.1,0.9,0.9])

# Create a pie chart of the number of Pokemon by type
types = ['Water', 'Normal', 'Flying', 'Grass', 'Psychic', 'Bug', 'Fire', 'Poison',
'Ground', 'Rock', 'Fighting', 'Dark', 'Steel', 'Electric', 'Dragon', 'Fairy',
'Ghost', 'Ice']
poke_num = [133, 109, 101, 98, 85, 77, 68, 66, 65, 60, 57, 54, 53, 51, 50, 50, 46, 40]

#https://matplotlib.org/2.0.2/examples/color/colormaps_reference.html
cmap = plt.cm.tab20c
colors = cmap(np.linspace(0., 1., len(types)))

explode = np.zeros(len(types))
explode[0] = 0.2


wedges, texts, autotexts = plt.pie(poke_num, 
                                   labels=types, 
                                   autopct='%1.0f%%', 
                                   startangle=90,
                                   colors=colors,
                                   explode=explode,
                                   shadow=True,
                                   textprops=dict(color="w"))

plt.legend(wedges, types, loc='right', bbox_to_anchor=(1, 0, 0.5, 1))
````
